```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv('//content//log_update.csv', encoding='ISO8859', error_bad_lines=False, header=None, sep=';', usecols=[0,1,2,4])
df.columns = ('date','time', 'id', 'event_type')
df
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S:%f')
```

```python
df.info()
```

```python
df_au = df[df['event_type'] == 'AU']
```

```python
df_au['day'] = df_au['date'].dt.day
df_au
```

```python
df_au.groupby('day')['id'].count().sort_values(ascending=False)
```

```python
df_au[df_au['day'] == 17]
```

```python
col_names = ['date','time', 'id', 'event_type']
func = lambda x: dt.datetime.strptime(x, '%d.%m.%Y %H:%M:%S:%f')
```

```python
df1 = pd.read_csv('//content//log_update.csv', encoding='ISO8859', error_bad_lines=False, header=None, sep=';', usecols=[0,1,2,4], names=col_names, 
                  parse_dates=[['date', 'time']], date_parser=func)
df1
```

```python
df1.info()
```

```python
df1.resample('5Min', on='date_time').id.count().sort_values(ascending=False)
```

```python
df1.resample('5Min', on='date_time', offset='1Min').id.count().sort_values(ascending=False)
```

```python
df1.resample('5Min', on='date_time', offset='2Min').id.count().sort_values(ascending=False)
```

```python
df1.resample('5Min', on='date_time', offset='3Min').id.count().sort_values(ascending=False)
```

```python
df1.resample('5Min', on='date_time', offset='4Min').id.count().sort_values(ascending=False)
```

```python
res = df1[df1['date_time'].dt.month == 2].groupby([pd.Grouper(key='date_time', freq='30Min'), 'event_type']).count()
res
```
